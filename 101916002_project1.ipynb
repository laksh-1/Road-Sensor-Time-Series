# Training and Testing (LGBM)

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor(random_state=51)
```

## Extract the test and training dataset for each section

```python
def extract_x_train(df, row_start, col_start):
  row_end = row_start + 9
  return df.iloc[row_start:row_end, 1:10]
```

```python
def extract_y_train(df, row_start, col_start):
  row_end = row_start + 9
  return df.iloc[row_start:row_end, col_start]
```

```python
def extract_x_test(df, row_start):
  row_end = row_start + 9
  return df.iloc[row_end, 1:10].to_numpy()
```

```python
def extract_y_test(df, row_start, col_start):
  row_end = row_start + 9
  return df.iloc[row_end, col_start]
```

```python
def calcuate_lmse(predict_val, test_val):
  pred_diff = predict_val - test_val
  return (pred_diff)**2
```

```python
column_name = ["Para-"+str(i) for i in range(9,14)]
XY = []
def getXYsplit(df, row_start, row_end):
  XY = []
  XY.append(extract_x_train(df, row_start, row_end))
  XY.append(extract_y_train(df, row_start, row_end))
  XY.append(extract_x_test(df,row_start))
  XY.append(extract_y_test(df,row_start,row_end))
  return XY
```

```python
predicted_values = defaultdict(lambda : [])
```

```python
target_col = [10,11,12,13,14]
root_mean_square_error = []
total_count = 0
curr_section = 0
total_squared_error = 0
max_val = 1000
for j in target_col:
  local_squared_error = 0
  l_count = 0
  LXY = [0,0,0,0]
  for i in range(0, max_val, 10):
    LXY = getXYsplit(df,i,j)
    x_train, y_train,x_test,y_test = LXY[0],LXY[1],LXY[2],LXY[3]
    model.fit(x_train, y_train)
    ypred = model.predict(x_test.reshape(1,-1))[0]
    predicted_values[column_name[curr_section]].append(ypred)
    total_squared_error = calcuate_lmse(ypred, y_test) + total_squared_error
    l_count += 1
    total_count += 1

  curr_section += 1
predicted_df = pd.DataFrame.from_dict(predicted_values)
```

```python

total_rmse_by_n = total_squared_error/total_count
final_rmse = total_rmse_by_n**0.5

print("The RMSE is " + str(final_rmse))
```
